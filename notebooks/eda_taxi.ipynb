```python
# importação de biblioteca, arquivo e validação de leitura
import duckdb

con = duckdb.connect()

preview = con.execute("""
    SELECT *
    FROM read_parquet('../data/raw/yellow_tripdata_2026-01.parquet')
    LIMIT 5
""").df()

preview
```

```python
# validação de schema
schema = con.execute("""
    DESCRIBE
    SELECT *
    FROM read_parquet('../data/raw/yellow_tripdata_2026-01.parquet')
""").df()

schema
```

```python
# validação de volume
total_linhas = con.execute("""
    SELECT COUNT(*) AS total_linhas
    FROM read_parquet('../data/raw/yellow_tripdata_2026-01.parquet')
""").df()

total_linhas
```
